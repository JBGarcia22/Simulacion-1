```python
import numpy as np
import matplotlib.pyplot as plt
import random as random

#Simulacion de una tirada:
def obtener_posicion(posicion = 1):
    #1: Azul    2: Rojo
    #3: Verde   4: Amarillo
    exitos = sum(np.random.randint(2, size=2))
    if posicion + exitos > 4:
        return 1
    else:
        return posicion + exitos

# Simulacion del juego
def experimento(posicion = 1, n = 1000):
    conteo = [0]*4
    posicion_actual = posicion
    posicion_nueva = 0
    for i in range(n):
        nueva_posicion = obtener_posicion(posicion_actual)
        conteo[nueva_posicion - 1] += 1
        posicion_actual = nueva_posicion
    return conteo

posicion_actual = 3
posicion_nueva = 0

n = 10000
juego = experimento(1, n)
print(juego)
plt.bar(range(1, 5), juego, tick_label=['Azul', 'Rojo', 'Verde', 'Amarillo'])
plt.xlabel('Posicion')
plt.ylabel('Cantidad de veces')
plt.title('Simulacion del juego')
plt.show()
```
