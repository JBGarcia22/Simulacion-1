# MÃ©todo de las 12 uniformes

```python
from random import random as rnd
import numpy as np
import matplotlib.pyplot as plt

#Devuelve la suma de 12 variables aleatorias uniformes
def twelve_unif():
    z = 0
    for i in range(12):
        z += rnd()
    return z

#Genera n variables aleatorias normales con el metodo de las 12 uniformes
#mu: media, sigma: desviaciÃ³n estÃ¡ndar
def normal(n = 1000, mu = 0, sigma = 1):
    x = []
    for i in range(n):
        x.append(mu + sigma * (twelve_unif() - 6))
    return x

mu = 5
sigma = 3
rand_values = normal(50000, mu, sigma)
plot = plt.hist(rand_values, bins = 100)
plt.title("MÃ©todo de las 12 uniformes")
plt.show()

print("Valor esperado:", np.mean(rand_values))
print("Valor esperado real:", mu)
print("Varianza:", np.var(rand_values))
print("Varianza real:", sigma**2)
```
