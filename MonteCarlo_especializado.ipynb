```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

def generate(n):
    x = []
    for i in range(n):
        R = rd.random()
        if (R < 1/2):
            x.append(np.sqrt(2*R))
        else:
            x.append(2*R)
    return x

val = generate(30000)

plot = plt.hist(val, bins = 500)
plt.show
```
