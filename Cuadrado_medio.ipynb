# Numeros Aleatorios

## MÃ©todo del medio del cuadrado

```python
from tabulate import tabulate as tab

#Se declara y define el valor de la semilla
x_i = 167359
n = len(str(x_i)) #Se obtiene el numero de digitos de la semilla. Se espera que dicho numero sea par

lista = [] #Conjunto donde se almacenaran los valores generados
lista.append(x_i)

x2 = ''

#Se genera una serie de valores aleatorios utilizando el metodo de cuarado medio
while True:
  #Se calcula el cuadrado del ultimo valor generado (o de la semilla si es la primera iteracion)
  x2 = str(x_i**2) #Se guarda como una cadena de caracteres para la manipulacion de sus digitos

  #Se anteponen 0s al numero calculado anterior si el numero de sus digitos es menor a 0
  #Despues se toma el valor generado por los n digitos centrales
  x_i = int(x2.zfill(2*n)[int(n/2):int(3*n/2)])

  #Si el nuevo valor generado es igual a alguno de los anteriores, la iteracion se detiene para evitar que los valores se ciclen
  if x_i not in lista:
    lista.append(x_i)
  else:
    break

#Se calculan los valores normalizados
lista_norm = [round(num/(10**n),n+1) for num in lista]
tabla = {"i" : range(len(lista)), "x_i" : lista, "x_i/m" : lista_norm}

#Se imprime una tabla con los valores generados junto con los normalizados, ademas del numero de valores generados
print(tab(tabla, headers = "keys",tablefmt="github"))

print("\nSe han generado ", len(lista), " valores")
```
