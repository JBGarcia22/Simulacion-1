## Metodo de Box-Muller


El método de Box-Muller simula $P(X,Y)$, donde $X,Y \thicksim N(0,1)$ e independientes, en coordenadas polares. Después calcula su representación en coordenadas cartesianas mediante

$$X = R\cos(\theta) \qquad Y = R\sin(\theta)$$

donde $R^2 \thicksim exp(\frac{1}{2})$ y $\theta \thicksim U(0,2\pi)$. Por lo que se puede usar el método de la transformada inversa para simular dichas variables aleatorias.

```python
from random import random as rnd
import numpy as np
import matplotlib.pyplot as plt

def Box_Muller():
    r = np.sqrt(-2*np.log(rnd()))
    theta = 2*np.pi*rnd()

    return r*np.cos(theta), r*np.sin(theta)

n = 30000

values = []

for i in range(n):
    values.extend(Box_Muller())

plt.hist(values, bins=100, alpha=0.6, color='g')
plt.title("Método de Box-Muller")

print("Valor esperado:", np.mean(values))
print("Valor esperado teórico:", 0)
print("Varianza:", np.var(values))
print("Varianza real:", 1)
```
