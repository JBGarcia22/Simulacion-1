```python
import numpy as np
from scipy.stats import uniform as runif
import matplotlib.pyplot as plt
```

# Metodo de Monte Carlo

```python
f = lambda x: np.sqrt(np.arctan(x))

X = runif.rvs(size = 10000)

I = sum(f(X))/len(X)

print(I)
```

```python
valores = []
j = 1000

for i in range(j):
  valores.append(sum(f(runif.rvs(size = i+1)))/float(i+1))

fig2,ax2 = plt.subplots(figsize=(20,5))
ax2.set_ylim([.4, 1])
ax2.set_xlim([0, j])
ax2.plot(valores, linewidth = .8, color="blue")
ax2.axhline(y=0.629823344207974, color='r', linestyle='-')
ax2.set_xlabel('Número de iteraciones')
ax2.set_ylabel('Valores generados')
ax2.grid(True)
```
