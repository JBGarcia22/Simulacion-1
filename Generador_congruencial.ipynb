## Generador Congruencial

```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate as tab

def generador (a,c,m,x_i) :
  return ((a*x_i + c)%m)

#Se declaran y definen los valores de los parametros junto con el valor de la semilla
a=1501          #multiplicador
c=1             #inclremento
m=10009         #modulo
x_0 = 57        #Semilla

#Se genera una lista de numeros aleatorios mediante el generador congruencial
#hasta que uno de los valores generados sea igual al de la semilla o se alcance el tamaño de ciclo maximo (m)
lista = [x_0]

x_i = x_0
for i in range(m) :
  x_i= generador(a,c,m,x_i)
  if (x_i == x_0):
    break
  else:
    lista.append(x_i)

#Se imprime una tabla con los valores generados junto con los mismos normalizados
#Ademas del tamaño del ciclo que se generó
lista_norm = [round(x/m,18) for x in lista] #Lista normalizada
tabla = {"i" : range(len(lista)), "x_i" : lista, "x_i/m" : lista_norm}

print(tab(tabla, headers = "keys",tablefmt="heavy_grid", floatfmt=".18f"))

print("\nLa lista contiene un ciclo de tamaño ", len(lista)) #Tamaño de la lista generada

#Se imprime el valor de la media y varianza de los valores normalizados, junto con un grafico de su distribucion
print("\nLa media es de ", np.mean(lista_norm),"\nLa varianza es de ", np.var(lista_norm))

fig, axs = plt.subplots(1,1,figsize=(10,5))
axs.hist(lista, bins=1000)
axs.set_title("Funcion de densidad")
plt.show()
```
